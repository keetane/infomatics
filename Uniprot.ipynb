```python
data = {
    'uniprotKB_accession' : ['Q6PML9', 'P61916', 'P04439', 'P01889', 'P10321', 'Q96SW2']
}
df = pd.DataFrame(data)
df
```

```python
uniprotIDs = df['uniprotKB_accession']
uniprotIDs
```

```python
#新しいカラムを作成
df["function"] = 0
df["subcellular location"] = 0
df["tissue specificity"] = 0
df
```

```python
url = 'https://www.uniprot.org/uniprot/Q6PML9.xml'
f = urlopen(url)
xml = f.read()
root = etree.fromstring(xml)
function = root.find('.//entry/name', root.nsmap)
print(function.text)
# print(etree.tostring(root, pretty_print=True).decode('UTF-8'))
```

```python
fullname = root.find('.//recommendedName/fullName',root.nsmap)
print(fullname.text)
shortname = root.find('.//recommendedName/shortName',root.nsmap)
print(shortname.text)
dbref = root.find('//dbreference[@type="PDB"]', root.nsmap).attrib
print(dbref)
```

```python
for type in root.iter('dbreference'):
    print('*** %s' % type.attrib)
```

```python
print(etree.tostring(root, pretty_print=True).decode())
```

```python
for uniprotID in uniprotIDs:
  url = "https://www.uniprot.org/uniprot/" + uniprotID + ".xml"
  f = urlopen(url)
  xml = f.read()
  root = etree.fromstring(xml)
  #以下のコードは下の説明を参照
  function = root.find('./entry/comment[@type="function"]', root.nsmap)
  if function==None:
    pass
  else:
    df["function"][uniprotIDs==uniprotID] = function[0].text

  sub_location = root.find('./entry/comment[@type="subcellular location"]', root.nsmap) 
  if sub_location==None: 
    pass 
  else: 
    df["subcellular location"][uniprotIDs==uniprotID] = sub_location.text 

  tissue = root.find('./entry/comment[@type="tissue specificity"]', root.nsmap)
  if tissue==None:
    pass
  else:
    df["tissue specificity"][uniprotIDs==uniprotID] = tissue[0].text

df
```

```python
df.to_excel('uniprot_Data.xlsx')
```

# UniprotKB accessionから情報を得る

```python
keyword = input('UniprotKB? ')
# CRBN = Q96SW2
# RNF43 = Q68DV7
```

```python
keys = [
    'Q96SW2', 
    'Q68DV7'
]
```

```python
import pandas as pd
from Bio import SeqIO
from Bio.PDB import PDBList as pdbl
from urllib.request import urlopen
```

```python
handle = urlopen("http://www.uniprot.org/uniprot/" + keyword + ".xml")
record = SeqIO.read(handle, "uniprot-xml")
# print(record)
```

```python
dbx = record.dbxrefs
dfIO = pd.DataFrame(dbx)
df = dfIO[0].str.split(':', expand=True)
df.columns = ['name', 'id', 'and']
df
```

```python
dfPDB = df[df['name']=='PDB']
PDBs = list(dfPDB['id'])
```

```python
data = {
    'name' : record.name, 
    'description' : record.description, 
    'pathway' : record.annotations['comment_pathway'], 
    'tissue' : record.annotations['comment_tissuespecificity'],
    'subcellularlocation' : ', '.join(record.annotations['comment_subcellularlocation_location']), 
    'function' : record.annotations['comment_function'],
    'PDB count' : str(len(PDBs)),
    'PDB' : ', '.join(PDBs)
}
df = pd.DataFrame(data, index = [record.id])
df.T
```

```python
for pdb_id in PDBs:
        pdbl().retrieve_pdb_file(pdb_id, pdir=keyword)
print('fertig')
```

```python
summary = pd.DataFrame(index = keys, columns = ['name', 'description', 'pathway', 'tissue', 'subcellularlocation', 'function', 'PDB count', 'PDB'])
summary
```

```python
summary['name']['Q96SW2']
```

```python
for keyword in keys:
    handle = urlopen("http://www.uniprot.org/uniprot/" + keyword + ".xml")
    record = SeqIO.read(handle, "uniprot-xml")
    dbx = record.dbxrefs
    dfIO = pd.DataFrame(dbx)
    df = dfIO[0].str.split(':', expand=True)
    df.columns = ['name', 'id', 'and']
    dfPDB = df[df['name']=='PDB']
    PDBs = list(dfPDB['id'])
    summary['name'][keyword] = record.name
    summary['description'][keyword] = record.description
    summary['pathway'][keyword] = ''.join(record.annotations['comment_pathway'])
    summary['tissue'][keyword] = ''.join(record.annotations['comment_tissuespecificity'])
    summary['subcellularlocation'][keyword] = ', '.join(record.annotations['comment_subcellularlocation_location'])
    summary['function'][keyword] = ''.join(record.annotations['comment_function'])
    summary['PDB count'][keyword] = str(len(PDBs))
    summary['PDB'][keyword] = ', '.join(PDBs)
summary.T
```

```python
summary.to_csv(keyword + '.csv')
```
